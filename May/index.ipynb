```python
import pandas as pd

# ==========================================
# LOAD EXCEL FILE
# ==========================================

file_path = r"maylines.xlsx"   # Change to your file

df = pd.read_excel(file_path)

# ==========================================
# CLEAN DATA
# ==========================================

df.columns = df.columns.str.strip()

df['Sim Serial Number'] = df['Sim Serial Number'].astype(str).str.strip()

df['Cumulative Usage'] = pd.to_numeric(
    df['Cumulative Usage'],
    errors='coerce'
)

df['Cumulative Commission'] = pd.to_numeric(
    df['Cumulative Commission'],
    errors='coerce'
)

# ==========================================
# PASTE SERIALS HERE
# ==========================================

serial_text = """
89254021344248461109
89254021344248460457
89254021344248460507
89254021344248460549
89254021344248461059
89254021344248460572
89254021344248461133
89254021344248461117

"""

# ==========================================
# EXTRACT SERIALS
# ==========================================

serial_numbers = [
    line.strip()
    for line in serial_text.splitlines()
    if line.strip()
]

# ==========================================
# CHECK SERIALS
# ==========================================

results = []

for serial in serial_numbers:

    match = df[df['Sim Serial Number'] == serial]

    if match.empty:
        results.append({
            "Serial Number": serial,
            "Status": "NOT FOUND",
            "Usage": "",
            "Commission": ""
        })
        continue

    row = match.iloc[0]

    usage = row['Cumulative Usage']
    commission = row['Cumulative Commission']

    quality = (
        pd.notna(usage)
        and pd.notna(commission)
        and usage >= 50
        and commission >= 300
    )

    results.append({
        "Serial Number": serial,
        "Status": "QUALITY" if quality else "NON QUALITY",
        "Usage": usage,
        "Commission": commission
    })

# ==========================================
# RESULTS
# ==========================================

results_df = pd.DataFrame(results)

print("\n===== RESULTS =====\n")
print(results_df.to_string(index=False))

# ==========================================
# SUMMARY
# ==========================================

quality_count = (results_df['Status'] == 'QUALITY').sum()
non_quality_count = (results_df['Status'] == 'NON QUALITY').sum()
not_found_count = (results_df['Status'] == 'NOT FOUND').sum()

print("\n===== SUMMARY =====")
print(f"Total Serials : {len(results_df)}")
print(f"Quality       : {quality_count}")
print(f"Non Quality   : {non_quality_count}")
print(f"Not Found     : {not_found_count}")
```

```python
import pandas as pd
import re

# ==========================================
# LOAD EXCEL
# ==========================================

file_path = r"maylines.xlsx"

df = pd.read_excel(file_path)

df.columns = df.columns.str.strip()

df['Sim Serial Number'] = (
    df['Sim Serial Number']
    .astype(str)
    .str.strip()
)

df['Cumulative Usage'] = pd.to_numeric(
    df['Cumulative Usage'],
    errors='coerce'
)

df['Cumulative Commission'] = pd.to_numeric(
    df['Cumulative Commission'],
    errors='coerce'
)

# ==========================================
# PASTE REPORT HERE
# ==========================================

report_text = """
Claypole trading ltd 
 31/05/2026 report RAVINE 
 Total lines activated -35 

 Rogers-4 
89254021344248458840
89254021344248458766
89254021344248458774
89254021344248458782

 BENEDITTE-9 
89254021344248454948
89254021344248454930
89254021344248454922
89254021344248454914
89254021344248454906
89254021344248454898
89254021344248454880
89254021344248454872
89254021344248454864

 Kiplimo-7 
89254021344248459244
89254021344248459236
89254021344248459228
89254021344248459210
89254021344248459202
89254021344248459194
89254021344248459186

 Kibet-5 
89254021344248454559
89254021344248454807
89254021344248454823
89254021344248454831
89254021344248454849

 Julia-10 
89254021344248455259
89254021344248455275
89254021344248455283
89254021344248455291
89254021344248455308
89254021344248455317
89254021344248455333
89254021344248455341
89254021344248458733
89254021344248455325
"""

# ==========================================
# EXTRACT DATE
# ==========================================

date_match = re.search(r'\d{2}/\d{2}/\d{4}', report_text)

report_date = (
    date_match.group()
    if date_match
    else "Unknown Date"
)

# ==========================================
# PARSE REPORT
# ==========================================

records = []

current_agent = None

for line in report_text.splitlines():

    line = line.strip()

    if not line:
        continue

    serial_match = re.fullmatch(r'\d{20}', line)

    if serial_match:

        records.append({
            'Agent': current_agent,
            'Serial Number': line
        })

        continue

    agent_match = re.match(
        r'^([A-Za-z ]+)\s*-\s*\d+$',
        line,
        re.IGNORECASE
    )

    if agent_match:
        current_agent = agent_match.group(1).strip()

# ==========================================
# CHECK SERIALS
# ==========================================

results = []

for rec in records:

    serial = rec['Serial Number']

    match = df[
        df['Sim Serial Number'] == serial
    ]

    if match.empty:

        results.append({
            'Date': report_date,
            'Agent': rec['Agent'],
            'Serial Number': serial,
            'Status': 'NOT FOUND'
        })

        continue

    row = match.iloc[0]

    usage = row['Cumulative Usage']
    commission = row['Cumulative Commission']

    quality = (
        pd.notna(usage)
        and pd.notna(commission)
        and usage >= 50
        and commission >= 300
    )

    results.append({
        'Date': report_date,
        'Agent': rec['Agent'],
        'Serial Number': serial,
        'Status': 'QUALITY' if quality else 'NON QUALITY'
    })

# ==========================================
# RESULTS
# ==========================================

results_df = pd.DataFrame(results)

print(f"\nREPORT DATE: {report_date}\n")

print(results_df.to_string(index=False))

# ==========================================
# SUMMARY BY AGENT
# ==========================================

print("\n" + "="*60)
print(f"REPORT DATE: {report_date}")
print("="*60)

for agent in sorted(results_df['Agent'].dropna().unique()):

    agent_df = results_df[
        results_df['Agent'] == agent
    ]

    total = len(agent_df)

    quality = len(
        agent_df[agent_df['Status'] == 'QUALITY']
    )

    non_quality_df = agent_df[
        agent_df['Status'] == 'NON QUALITY'
    ]

    not_found_df = agent_df[
        agent_df['Status'] == 'NOT FOUND'
    ]

    non_quality = len(non_quality_df)
    not_found = len(not_found_df)

    non_quality_serials = [
        str(x)[-5:]
        for x in non_quality_df['Serial Number']
    ]

    not_found_serials = [
        str(x)[-5:]
        for x in not_found_df['Serial Number']
    ]

    print(f"\n{agent.upper()}")
    print("-" * 40)

    print(f"Total Submitted : {total}")
    print(f"Quality         : {quality}")
    print(f"Non Quality     : {non_quality}")
    print(f"Not Found       : {not_found}")

    print("\nNon Quality Serials:")
    print(
        ", ".join(non_quality_serials)
        if non_quality_serials
        else "None"
    )

    print("\nNot Found Serials:")
    print(
        ", ".join(not_found_serials)
        if not_found_serials
        else "None"
    )

    print()
```
