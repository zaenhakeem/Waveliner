```python

# Load the first sheet
df = pd.read_excel(file_path)

# Preview the data
print("First 5 rows:")
display(df.head())
```

```python

# Dataset information
print("\nDataset Info:")
print(df.info())

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python

# Basic statistics
print("\nSummary Statistics:")
display(df.describe(include='all'))

# Number of rows and columns
print("\nShape of dataset:")
print(df.shape)

# -----------------------------
# Example Analysis
# -----------------------------

# Count unique values per column
print("\nUnique values per column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")

# If you have a County column
if 'County' in df.columns:
    print("\nCounty Counts:")
    display(df['County'].value_counts())

# Filter Nairobi rows
if 'County' in df.columns:
    nairobi_df = df[df['County'] == 'Nairobi']
    
    print("\nNairobi Records:")
    display(nairobi_df.head())

# -----------------------------
# Visualization
# -----------------------------

# Plot top counties if County exists
if 'County' in df.columns:
    plt.figure(figsize=(10,6))
    df['County'].value_counts().head(10).plot(kind='bar')
    plt.title("Top Counties")
    plt.xlabel("County")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

# Correlation heatmap for numeric columns
numeric_df = df.select_dtypes(include='number')

if not numeric_df.empty:
    plt.figure(figsize=(10,6))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
    plt.title("Correlation Heatmap")
    plt.show()
```

```python
# ====================================
# REGION COMPOSITION ANALYSIS
# ====================================

import pandas as pd
import matplotlib.pyplot as plt

# Read file
df = pd.read_excel("mpesa.xlsx")

# ------------------------------------
# REGION COMPOSITION
# ------------------------------------

print("Unique Regions:")
display(df['Region'].unique())

# Count records per region
region_counts = df['Region'].value_counts()

print("\nRegion Composition Counts:")
display(region_counts)

# Percentage composition
region_percentage = (
    df['Region'].value_counts(normalize=True) * 100
).round(2)

print("\nRegion Composition Percentage:")
display(region_percentage)

# Combine counts + percentages
region_summary = pd.DataFrame({
    'Count': region_counts,
    'Percentage (%)': region_percentage
})

print("\nRegion Summary")
display(region_summary)

# ------------------------------------
# VISUALIZATION
# ------------------------------------

# Bar Chart
plt.figure(figsize=(12,6))

region_counts.plot(
    kind='bar',
    color='skyblue'
)

plt.title("Region Composition")
plt.xlabel("Region")
plt.ylabel("Number of Records")
plt.xticks(rotation=45)

plt.show()

# ------------------------------------
# PIE CHART
# ------------------------------------

plt.figure(figsize=(8,8))

region_counts.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel("")
plt.title("Region Composition Percentage")

plt.show()

# ------------------------------------
# TERRITORY COMPOSITION WITHIN REGIONS
# ------------------------------------

region_territory = pd.crosstab(
    df['Region'],
    df['Territory']
)

print("\nTerritory Composition Per Region")
display(region_territory)

# ------------------------------------
# CLUSTER COMPOSITION WITHIN REGIONS
# ------------------------------------

region_cluster = pd.crosstab(
    df['Region'],
    df['Cluster']
)

print("\nCluster Composition Per Region")
display(region_cluster)
```

```python
# ==========================================
# TERRITORY COMPOSITION PER REGION (WRITTEN)
# ==========================================

import pandas as pd

# Read the Excel file
df = pd.read_excel("mpesa.xlsx")

# Remove missing values if any
df = df.dropna(subset=['Region', 'Territory'])

# Group territories by region
region_territories = (
    df.groupby('Region')['Territory']
    .unique()
)

# Print in written form
for region, territories in region_territories.items():
    
    territory_list = ", ".join(sorted(territories))
    
    print(f"\n{region} region is composed of the following territories:")
    print(f"{territory_list}")
    print("-" * 80)
```

```python
# ===================================================
# TERRITORY COMPOSITION PER REGION WITH NUMBERS
# ===================================================

import pandas as pd

# Read Excel file
df = pd.read_excel("mpesa.xlsx")

# Remove missing values
df = df.dropna(subset=['Region', 'Territory'])

# Group by Region and Territory
grouped = (
    df.groupby(['Region', 'Territory'])
    .size()
    .reset_index(name='Count')
)

# Print in written form
for region in grouped['Region'].unique():
    
    print(f"\n{region} region is composed of:")
    
    region_data = grouped[grouped['Region'] == region]
    
    for _, row in region_data.iterrows():
        territory = row['Territory']
        count = row['Count']
        
        print(f" - {territory}: {count} records")
    
    print("-" * 80)
```
